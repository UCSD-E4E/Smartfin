## Get ride from api
this is in order to get a location, start and end date. Alternatively you can choose your own start and end dates, and your own CDIP buoy station too. 

```python
rideId = '15692'
rideGetSingle = f'http://ec2-54-203-7-235.us-west-2.compute.amazonaws.com/ride/rides/rideId={rideId}' # creates new ride and upload all ride data to db
response = requests.get(rideGetSingle).json()
response
```

```python
startTime = response['startTime']
endTime = response['endTime']
buoyCDIP = response['buoyCDIP']
heightCDIP = response['heightCDIP']
```
